```python
import random

successors = {'A': [['T', 11], ['B', 13], ['C', 21]], 'T': [['D', 27], ['B', 13]], 'B': [['D', 27], ['E', 3]],
              'C': [['F', 25], ['G', 4]], 'D': [['H', 101], ['I', 99]], 'F': [['J', 67]], 'G': [['K', 99], ['L', 3]],
              'H': [['M', 17]], 'I': [['M', 17]], 'J': [['M', 17]]}

start_node = 'A'
closed_list = list()
candidate_list = [['T', 11], ['B', 13], ['C', 21]]


def move_generator(node):
    new_list = list()
    if node in successors.keys():
        new_list = successors[node]
    return new_list


def sort_nodes(node_list):
    node_list.sort(key=lambda x: x[1], reverse=True)
    return node_list


def heuristic(node):
    return node[1]


def append_lists(list1, list2):
    new_list = list(list1) + list(list2)
    return new_list


def iterated_hill_climbing(start, iterations):
    global closed_list
    current_node = [start, 5]
    best_node = current_node
    closed_list = [current_node]
    print("\nInitial Point =", current_node)

    for i in range(1, iterations + 1):
        print("\n*******************************")
        print("For iteration i =", i)
        current_node = random.choice(candidate_list)
        children = move_generator(current_node[0])
        sort_nodes(children)
        print("\nStart =", current_node)
        print("Sorted Child List =", children)
        new_node = children[0]
        closed_list = [current_node]

        while heuristic(new_node) > heuristic(current_node):
            print("\n---------")
            current_node = new_node
            print("Current Node =", current_node)
            closed_list = append_lists(closed_list, [current_node])
            children = move_generator(current_node[0])
            print("Child List =", children)
            sort_nodes(children)
            print("Sorted Child List =", children)
            print("Closed List =", closed_list)
            new_node = children[0]
            if heuristic(new_node) > heuristic(best_node):
                best_node = new_node

    return best_node


best_node = iterated_hill_climbing(start_node, 3)  # call search algorithm
print("Best Node =", best_node)
```
